Slicing

s[a:b:c]

```python
s = "I love Bangkok!"
# "I love" --> 0,1,2,3,4,5 -> range (0,6)
print(s[0:6]) 

# "Bangkok" -> 7,8,9,10,11,12,13 -> range(7, 14)
print(s[7:14]) 

print(s[1:8:2]) # 1,3,5,7 -> "oeB"
print(s[-3:-10:-2]) #-3, -5, -7, -9 -> "oga"
```
